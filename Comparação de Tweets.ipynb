# Comparação
    A seguir será mostrada uma comparação de Posts, Favoritados e Retweets de cada politico.
    Foi usado o grafico de barras para melhor visualização.
    matplotlib.pylot.bar(https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.bar.html)
    matplotlib.pylot.annotate(https://matplotlib.org/3.3.3/api/_as_gen/matplotlib.pyplot.annotate.html)

```python
x = np.arange(len(temas))
```

```python
largura = 0.35
```

```python
def autolabel(grupos):
    for i in grupos:
        h = i.get_height()
        ax.annotate('{}'.format(h),
                    xy = (i.get_x()+i.get_width()/2,h),
                    xytext = (0,3),
                    textcoords = 'offset points',
                    ha = 'center')
```

```python
fig,ax = plt.subplots(figsize = (15,6))

posts_bolsonaro = ax.bar(x -largura/2,valores_posts_bolsonaro,largura,label = 'Bolsonaro')
posts_lula = ax.bar(x +largura/2,valores_posts_lula,largura,label = 'Lula')

ax.set_title('Pontuação por Posts')
ax.set_ylabel('Pontuação')

ax.legend()
ax.set_ylim([0, 400])

ax.set_xticks(x)

ax.set_xticklabels(temas)

autolabel(posts_bolsonaro)
autolabel(posts_lula)

plt.show()
```

```python
fig,ax = plt.subplots(figsize = (15,6))

favoritos_bolsonaro = ax.bar(x -largura/1.5,valores_favoritos_bolsonaro,largura,label = 'Bolsonaro')
favoritos_lula = ax.bar(x +largura/1.5,valores_favoritos_lula,largura,label = 'Lula')

ax.set_title('Pontuação por favoritos')
ax.set_ylabel('Pontuação')

ax.legend()
ax.set_ylim([0, 4200000])

ax.set_xticks(x)

ax.set_xticklabels(temas)

autolabel(favoritos_bolsonaro)
autolabel(favoritos_lula)

plt.show()
```

```python
fig,ax = plt.subplots(figsize = (15,6))

retweet_bolsonaro = ax.bar(x -largura/1.5,valores_retweet_bolsonaro,largura,label = 'Bolsonaro')
retweet_lula = ax.bar(x +largura/1.5,valores_retweet_lula,largura,label = 'Lula')

ax.set_title('Pontuação por retweet')
ax.set_ylabel('Pontuação')

ax.legend()
ax.set_ylim([0, 620000])

ax.set_xticks(x)

ax.set_xticklabels(temas)

autolabel(retweet_bolsonaro)
autolabel(retweet_lula)

plt.show()
```

# Considerações Finais
   <H2> Ideologia Politica </H2>
   Ao analisar os twittes dos politicos podemos observar que os posts do Lula são basicamente voltado a educação, cultura, familia... isso são traços de uma ideologia politica de esquerda. (https://www.significados.com.br/esquerda/)
<br>
   Por sua vez os twittes do bolsonaro são mais voltado a Segurança, familia, corrupção... que sao traços de uma ideologia politica de direita(https://www.significados.com.br/direita/)
   
   <H2> Repercução </H2>
   
   Após analisar a repercução dos politicos foi visto que mesmo com poucos twittes sobre os assuntos propostos o Bolsonaro em todos os casos se destacou bastante.
